```python
import numpy as np
import matplotlib as plt
import pandas
```

```python
def load():
    data = np.loadtxt('./ex2data1.txt', delimiter=',')
    # print(data[:,2],data[:,:2])
    x= data[:,:2]
    y = data[:,2]
    return x, y   
```

```python
x_train, y_train = load()
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python

plt.pyplot.plot(x_train[y_train==1, 0], x_train[y_train==1, 1], 'k+', label='Got Admission')
plt.pyplot.plot(x_train[y_train==0, 0], x_train[y_train==0, 1], 'yo', label='No admission')
plt.pyplot.xlabel('Exam 1')
plt.pyplot.ylabel('Exam 2')
plt.pyplot.legend(loc='upper right')
plt.pyplot.show()
```
